### Iteration in Numpy array

```python
x = np.array([9,10,11,12])
for i in x:
    print (i)
```

```python
x = np.array([[9,10,11,12],[3,4,5,6]])
for i in x:
    for j in i:
        print(j)
    
```

```python
x = np.array([[[9,10,11,12],[3,4,5,6],[0,1,2,3]]])
for i in x:
    for j in i:
        for k in j:
            print(k)

print()
print(x.ndim)
       
```

### we have fxn nditer()

```python
x = np.array([[[9,10,11,12],[3,4,5,6],[0,1,2,3]]])
for i in np.nditer(x):
    print(i)
```

### to find index and iteration ndenumerate()

```python
x = np.array([[[9,10,11,12],[3,4,5,6],[0,1,2,3]]])
for i,d in np.ndenumerate(x):
    print(i,d)
```

## view vs copy

```python
var =np.array([1,2,3,4])
co = var.copy()
var[2]=50
print(var)
print(co)
```

```python
var =np.array([1,2,3,4])
viw = var.view()
var[2]=50
print(var)
print(viw)
```

### joining array using fxn

### concatenate()

```python
var1 = np.array([1,2,3])
var2 = np.array([5,6,7])
con = np.concatenate((var1,var2))
print(con)
```

```python
var1 = np.array([[1,2,3],[11,12,13]])
var2 = np.array([[5,6,7],[15,16,17]])
con = np.concatenate((var1,var2))
print(con)
print()
con2 = np.concatenate((var1,var2),axis=0)
print(con2)
```

```python
var1 = np.array([[1,2,3],[11,12,13]])
var2 = np.array([[5,6,7],[15,16,17]])
print(var1)
print(var2)
print()

con1 = np.stack((var1,var2),axis=1)
print(con1)
```

```python
var1 = np.array([[1,2,3],[11,12,13]])
var2 = np.array([[5,6,7],[15,16,17]])
# print(var1)
# print(var2)
# print()
con1 = np.vstack((var1,var2)) #vertical

con2 = np.hstack((var1,var2)) #horizontal

con3 = np.dstack((var1,var2)) #height


print(con1)
print()
print(con2)
print()
print(con3)
```

## Functions in Numpy
###  Search


```python
var = np.array([1,2,3,2,5,2])
#index          0,1,2,3,4,5
s = np.where(var == 2)
print(s)
```

```python
var = np.array([2,3,4,5,6,7])
#index          0,1,2,3,4,5
s = np.where(var%2 == 0)
print(s)
```

### sort

```python
var = np.array([22,3,42,5,36,7])
print(np.sort(var))
```

### Searchsorted

```python
var = np.array([1,2,3,5,6])
# index         0,1,2,3,4
ss = np.searchsorted(var,4)
print(ss)
```

```python
var = np.array([1,2,3,5,6])
# index         0,1,2,3,4
ss = np.searchsorted(var,4,side='right')
print(ss)
```

### flatten and ravel = covert 2d to 1d`

```python
var = np.array([1,2,3,4,5,6])
x = np.resize(var,(2,3))
print(x)
print(x.flatten())
print(x.flatten(order = 'F'))
print(np.ravel(x,order = 'c'))
```
