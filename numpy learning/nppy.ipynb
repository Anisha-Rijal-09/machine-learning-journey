```python
x=[1,2,3,4]
y=np.array(x)
print(y)
print(y.ndim)
```

```python
y=np.array([[1,2,3,4],[5,6,7,8]])
print(y)
print(y.ndim)
```

```python
arn = np.array([1,2,3,4],ndmin=10)
print(arn)
```
