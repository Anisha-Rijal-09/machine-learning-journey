### Annotations #place text in graph

```python
data = pd.read_csv("C:\\Users\\sampada\\machine-learning-journey\\data  visualization\\batter.csv" )
```

```python
d = data.head(100).sample(50,random_state=5)
d
```

```python
plt.figure(figsize=(20,10))
plt.scatter(d['avg'],d['strike_rate'])
for i in range(d.shape[0]):
    plt.text(d['avg'].values[i],d['strike_rate'].values[i],d['batter'].values[i])
```
