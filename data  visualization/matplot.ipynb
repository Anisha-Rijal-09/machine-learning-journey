## 3-D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
 xx,yy =np.meshgrid(x,y)

```

```python
z= xx**2 +yy**2
```

```python
z.shape
```

```python
fig =plt.figure(figsize=(10,7))
ax =plt.subplot(projection = '3d')
c = ax.plot_surface(xx,yy,z,cmap = 'viridis')
fig.colorbar(c)
```

## contour plot -3d to 2d

```python
fig =plt.figure(figsize=(10,7))
ax =plt.subplot()
c =ax.contour(xx,yy,z,cmap = 'viridis')
fig.colorbar(c)
```

```python
fig =plt.figure(figsize=(10,7))
ax =plt.subplot()
c =ax.contourf(xx,yy,z,cmap = 'viridis')
fig.colorbar(c)
```

## Heatmap

```python
hd = pd.read_csv("C:\\Users\\sampada\\machine-learning-journey\\data  visualization\\football_goals.csv")
```

```python
hd.head()
```

```python


# Create time intervals
bins = [0, 15, 30, 45, 60, 75, 90]
labels = ["0-15", "16-30", "31-45", "46-60", "61-75", "76-90"]

hd["Time_Period"] = pd.cut(
    hd["Goal_Minute"],
    bins=bins,
    labels=labels,
    include_lowest=True
)

# Create pivot table
heatmap_data = pd.crosstab(hd["Player"], hd["Time_Period"])

# Plot heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(heatmap_data, annot=True, cmap="YlGnBu")
plt.title("Goals Scored by Players in Different Time Periods")
plt.xlabel("Match Time (Minutes)")
plt.ylabel("Player")
plt.show()
```
