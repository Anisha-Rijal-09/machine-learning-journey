## Series

```python
import pandas as pd
```

```python
x = [1,2,3,4,5]
var = pd.Series(x)
print(var)
print(type(var))
```

```python
x = [3,4,5]
var = pd.Series(x,index = ['a','b','c'],dtype = float,name = 'python')
print(var)



```

```python
dict = {'name':['pytho','java','c','c++'],'rank':[1,2,4,3]}
var = pd.Series(dict)
print(var)
```

```python
x = pd.Series(12)
print(x)
```

```python
x = pd.Series(12,index = [1,2,3,4,5])
print(x)
```

```python
s1 = pd.Series(12,index = [1,2,3,4,5])
s2 = pd.Series(12,index = [3,4,5])
print(s1*s2)
```

### DataFrames

```python
import pandas as pd
```

```python
d = {'a':[1,2,3,4,5],'b':[11,12,13,14,15]}
var = pd.DataFrame(d)
print(type(var))
print(var)
```

```python
d = {'a':[1,2,3,4,5],'b':[11,12,13,14,15]}
var = pd.DataFrame(d,index=['a','s','d','f','g'],columns=['a'])
print(type(var))
print(var)
print(var['a']['f'])
```

```python
d = {'a':[1,2,3,4,5],'b':[11,12,13,14,15]}
var = pd.DataFrame(d)
print(var)
print(var['b'][2])
```

```python
ls = [[1,2,3,4],[21,22,23,24]] # list of list
var =pd.DataFrame(ls)
print(type(var))
print(var)
```

```python
list_1 = {'a':pd.Series([1,2,3,4]),'b':pd.Series([31,32,33,34])}
var = pd.DataFrame(list_1)
print(var)
print(type(var))
```

### Airthmetic operation in pandas

```python
var_1 = pd.DataFrame({'a':[1,11,21,31],'b':[10,20,30,40]})
var_1
```

```python

var_1['c'] = var_1['a']+var_1['b']
var_1['d'] = var_1['a']-var_1['b']
var_1['e'] = var_1['a']*var_1['b']
var_1['f'] = var_1['a']/ var_1['b']
var_1
```

```python
var_1['new']=var_1['b']<= 20
var_1
```
