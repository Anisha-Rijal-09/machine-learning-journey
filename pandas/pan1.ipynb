## Merge


```python
import pandas as pd
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,4],'c':[21,22,23,24]})
pd.merge(var1,var2,on ='A')
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,5],'c':[21,22,23,24]})
pd.merge(var1,var2,how ='left')
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,5],'c':[21,22,23,24]})
pd.merge(var1,var2,how ='right')
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,5],'c':[21,22,23,24]})
pd.merge(var1,var2,how ='right',indicator=True)
```

```python
pd.merge(var1,var2,how ='left',indicator=True)
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,4],'B':[21,22,23,24]})
pd.merge(var1,var2,left_index = True,right_index = True)#when same name 
```

```python
pd.merge(var1,var2,left_index = True,right_index = True,suffixes=('name','id'))

```

## Concat

```python
s1= pd.Series([1,2,3,4])
s2= pd.Series([21,22,23,24])
pd.concat([s1,s2])
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'A':[1,2,3,5],'B':[21,22,23,24]})
pd.concat([d1,d2])
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'A':[1,2,3,5],'B':[21,22,23,24]})
pd.concat([d1,d2],axis=1)
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'A':[1,2],'B':[21,22]})
pd.concat([d1,d2],axis=1,join='inner')#intersection
```

```python
pd.concat([d1,d2],axis=1,join='outer')#union
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'A':[1,2,3,5],'c':[21,22,23,24]})
pd.concat([d1,d2],axis=1,keys=['d1','d2'])
```

```python
pd.concat([d1,d2],axis=0,keys=['d1','d2'])
```

## GroupBY


```python
var = pd.DataFrame({'name':['a','b','c','a','c','b','c','a','b','a'],
                    's1':[12,13,14,23,12,15,13,12,24,23],
                    's2':[22,23,24,22,12,24,23,12,22,24]})
var
```

```python
var_new  = var.groupby('name')
var_new
```

```python
for x,y in var_new:
    print(x)
  
    print(y)
    print()
```
