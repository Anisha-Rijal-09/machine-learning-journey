## Merge


```python
import pandas as pd
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,4],'c':[21,22,23,24]})
pd.merge(var1,var2,on ='A')
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,5],'c':[21,22,23,24]})
pd.merge(var1,var2,how ='left')
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,5],'c':[21,22,23,24]})
pd.merge(var1,var2,how ='right')
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,5],'c':[21,22,23,24]})
pd.merge(var1,var2,how ='right',indicator=True)
```

```python
pd.merge(var1,var2,how ='left',indicator=True)
```

```python
var1 =pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,14]})
var2 =pd.DataFrame({'A':[1,2,3,4],'B':[21,22,23,24]})
pd.merge(var1,var2,left_index = True,right_index = True)#when same name 
```

```python
pd.merge(var1,var2,left_index = True,right_index = True,suffixes=('name','id'))

```

## Concat

```python
s1= pd.Series([1,2,3,4])
s2= pd.Series([21,22,23,24])
pd.concat([s1,s2])
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'A':[1,2,3,5],'B':[21,22,23,24]})
pd.concat([d1,d2])
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'A':[1,2,3,5],'B':[21,22,23,24]})
pd.concat([d1,d2],axis=1)
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'A':[1,2],'B':[21,22]})
pd.concat([d1,d2],axis=1,join='inner')#intersection
```

```python
pd.concat([d1,d2],axis=1,join='outer')#union
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'A':[1,2,3,5],'c':[21,22,23,24]})
pd.concat([d1,d2],axis=1,keys=['d1','d2'])
```

```python
pd.concat([d1,d2],axis=0,keys=['d1','d2'])
```

## GroupBY


```python
var = pd.DataFrame({'name':['a','b','c','a','c','b','c','a','b','a'],
                    's1':[12,13,14,23,12,15,13,12,24,23],
                    's2':[22,23,24,22,12,24,23,12,22,24]})
var
```

```python
var_new  = var.groupby('name')
var_new
```

```python
for x,y in var_new:
    print(x)
  
    print(y)
    print()
```

```python
var_new.get_group('a')
```

```python
var_new.min()
```

## join

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'C':[1,2,3,5],'D':[21,22,23,24]})
d1.join(d2)
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'C':[1,2],'D':[21,22]})
d1.join(d2)
```

```python
d1.join(d2,how = 'inner')
```

```python
d1.join(d2,how = 'outer')
```

```python
d1.join(d2,how = 'left')
```

```python
d1.join(d2,how = 'right')
```

```python
d1= pd.DataFrame({'A':[1,2,3,4],'B':[11,12,13,24]})
d2 = pd.DataFrame({'A':[1,2],'B':[21,22]})# column name same then value eror us raised to remove this suffix is used
d1.join(d2,how = 'left',lsuffix = '_12')
```

```python
d1.join(d2,rsuffix = '_12')
```

## Reshaping using melt() n pivot()


### melt()

```python
var = pd.DataFrame({'Days':[1,2,3,4,5,6],'eng':[11,12,13,14,15,16],'sci':[17,18,19,20,12,23]})
var
```

```python
pd.melt(var)# represnt data vertivally
```

```python
pd.melt(var,id_vars=['Days'])
```

```python
pd.melt(var,id_vars=['eng'],var_name ='python')
```

```python
pd.melt(var,id_vars=['eng'],var_name ='python',value_name = 'val')
```

### pivot()

```python
var = pd.DataFrame({'Days':[1,2,3,4,5,6],'st_name':['a','b','c','a','b','c'],'eng':[11,12,13,14,15,16],'sci':[17,18,19,20,12,23]})
var
```

```python
var.pivot(index='Days',columns = 'st_name')
```

```python
var.pivot(index='Days',columns = 'st_name',values = 'eng')# to get only one column data
```

```python
var = pd.DataFrame({'Days':[1,1,1,1,2,2],'st_name':['a','b','c','a','b','c'],'eng':[11,12,13,14,15,16],'sci':[17,18,19,20,12,23]})
var
```

```python
var.pivot_table(index = 'st_name',columns ='Days',aggfunc= 'mean')
```

```python
var.pivot_table(index = 'st_name',columns ='Days',aggfunc= 'sum')
```

```python
var.pivot_table(index = 'st_name',columns ='Days',aggfunc= 'sum',margins =True)# margin gives mean
```
