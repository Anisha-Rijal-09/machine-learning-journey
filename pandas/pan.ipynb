## Series

```python
import pandas as pd
```

```python
x = [1,2,3,4,5]
var = pd.Series(x)
print(var)
print(type(var))
```

```python
x = [3,4,5]
var = pd.Series(x,index = ['a','b','c'],dtype = float,name = 'python')
print(var)



```

```python
dict = {'name':['pytho','java','c','c++'],'rank':[1,2,4,3]}
var = pd.Series(dict)
print(var)
```

```python
x = pd.Series(12)
print(x)
```

```python
x = pd.Series(12,index = [1,2,3,4,5])
print(x)
```

```python
s1 = pd.Series(12,index = [1,2,3,4,5])
s2 = pd.Series(12,index = [3,4,5])
print(s1*s2)
```

### DataFrames

```python
import pandas as pd
```

```python
d = {'a':[1,2,3,4,5],'b':[11,12,13,14,15]}
var = pd.DataFrame(d)
print(type(var))
print(var)
```
