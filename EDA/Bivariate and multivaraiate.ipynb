## Numerical-Numerical -> Scatter plot

## NUmerical-Categorical -> barplot

```python
sns.barplot(data=df,x='Pclass',y='Age')
```

```python
sns.barplot(data=df,x='Pclass',y='Age',hue='Sex')
```

## NUmerical-Categorical -> boxplot

```python
sns.boxplot(data=df,x='Sex',y='Age')
```

```python
sns.boxplot(data=df,x='Sex',y='Age',hue='Survived')
```

## NUmerical-Categorical -> distplot

```python
sns.distplot(df[df['Survived']==0]['Age'])
```

```python
sns.distplot(df[df['Survived']==0]['Age'],hist=False)
sns.distplot(df[df['Survived']==1]['Age'],hist=False)
```
